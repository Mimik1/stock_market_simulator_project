```python
data= pd.merge(oil[['Date','Close']], aapl[['Date','Close']], on='Date')
#data= pd.merge(data, gold[['Date','Close']], on='Date')
#data.columns=['date','oil','aapl','gold']
data.columns=['date','oil','aapl']
```

```python
data=data.dropna()
```

```python
#data=data[['aapl','oil','gold']].values
data=data[['aapl','oil']].values
```

```python
data[:,0]=data[:,0]/500
data[:,1]=data[:,1]/500
#data[:,2]=data[:,2]/2000
```

```python
x=[]
y=[]
for i in range(3,data.shape[0]):
    x.append([*data[i-3],*data[i-2],*data[i-1]])
    y.append(data[i,0])
x=np.reshape(np.array(x),(-1,1,6))#9
y=np.reshape(np.array(y),(-1,1))
```

```python
model=tf.keras.Sequential()
model.add(tf.keras.layers.LSTM(10,input_shape=(x.shape[1],x.shape[2])))
model.add(tf.keras.layers.Dense(1))
model.compile(loss='mae',optimizer='adam')
for i in range(600):
    a=np.random.randint(2989)
    b=np.random.randint(2989)
    if a>b:
        a,b=b,a
    model.fit(x[a:b,:,:],y[a:b,:],epochs=20,batch_size=1,shuffle=False)
    model.save(str(i)+'.h5')
    print(i)
```
